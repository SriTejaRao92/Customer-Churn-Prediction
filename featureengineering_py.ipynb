```python
!pip install scikit-learn
import pandas as pd
from sklearn.utils import resample # Import the resample function
from sklearn.model_selection import train_test_split # Import train_test_split
import lightgbm as lgb
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
encoded_df = pd.DataFrame({'PhoneService': [1, 0, 1, 0, 1],
                           'MultipleLines': [0, 1, 1, 0, 1],
                           'InternetService': [1, 1, 0, 1, 0],
                           'OnlineSecurity': [0, 1, 0, 1, 0],
                           'OnlineBackup': [1, 0, 1, 0, 1],
                           'DeviceProtection': [0, 1, 0, 0, 1],
                           'TechSupport': [1, 0, 1, 1, 0],
                           'StreamingTV': [0, 1, 0, 1, 1],
                           'StreamingMovies': [1, 0, 1, 0, 0],
                           'PaperlessBilling': [0, 1, 1, 1, 0],
                           'Churn': [0, 1, 0, 1, 0],  # Add the 'Churn' column
                           'tenure': [12, 36, 24, 1, 48],  # Example tenure data
                           'MonthlyCharges': [65.5, 85.0, 50.0, 75.0, 90.0], # Example MonthlyCharges data
                           'TotalCharges': [800.0, 3000.0, 1200.0, 75.0, 4300.0], # Example TotalCharges data
                           'customerID': [1, 2, 3, 4, 5],  # Example customer IDs
                           'gender': [0, 1, 0, 1, 0],  # Example gender data (assuming binary encoding)
                           'SeniorCitizen': [0, 1, 0, 0, 1] # Example SeniorCitizen data
                          })
services = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
            'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling']
encoded_df['services'] = encoded_df[services].apply(lambda x: x.sum(), axis=1)
encoded_df.head()
features = encoded_df.drop(columns=['Churn','tenure','customerID','gender',
                                    'SeniorCitizen','PhoneService']).columns
target = ['Churn']
df_majority = encoded_df[encoded_df.Churn == 0]
df_minority = encoded_df[encoded_df.Churn == 1]

# You'll need to define or import 'down' here, as it's used in the resampling step below
down = len(df_minority)  # Example: set 'down' to the size of the minority class


# Downsample majority class
df_majority_downsampled = resample(df_majority,
                                 replace=False,    # sample without replacement
                                 n_samples=down,     # to match minority class
                                 random_state=42) # reproducible results
# combine the new dataframes
df_downsampled = pd.concat([df_majority_downsampled, df_minority])

df_downsampled.Churn.value_counts()
y = df_downsampled[target]
X = df_downsampled[features]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50, random_state=42, stratify=y)
#LightGBM
clf = lgb.LGBMClassifier(drop_rate=0.9, min_data_in_leaf=800, max_bin=500,
                         n_estimators=5000, min_sum_hessian_in_leaf=1, importance_type='gain',
                         learning_rate=0.4, bagging_fraction=0.9, colsample_bytree=1.0,
                         feature_fraction=0.1, lambda_l1=5.0, lambda_l2=3.0, max_depth=9,
                         min_child_samples=55, min_child_weight=5.0, min_split_gain=0.1,
                         num_leaves=4000, subsample = 0.75)

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(f'Precision: {precision_score(y_test, y_pred)}')
print(f'Recall: {recall_score(y_test, y_pred)}')
print(f'Confusion Matrix: {confusion_matrix(y_test, y_pred)}')

fig, ax = plt.subplots(figsize = (10,10))
sns.set_style(style='whitegrid')
y_prob = clf.predict_proba(X_test)[::,1]
y_pred_proba = y_prob
fpr_rand, tpr_rand, _ = roc_curve(y_test,  y_pred_proba)
auc_rand = roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr_rand,tpr_rand,label="LightGBM, AUC="+str(auc_rand))
plt.legend(loc=4)
plt.title('ROC Curve')
plt.show()

feature_imp = pd.DataFrame(sorted(zip(clf.feature_importances_, X.columns)), columns=['Value','Feature'])
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False))
plt.title('LightGBM Features (avg over folds)')
plt.tight_layout()
plt.show()
encoded_df['avg_price'] = encoded_df.MonthlyCharges / encoded_df.services
encoded_df.head()
features = encoded_df.drop(columns=['Churn','tenure','customerID','gender','services',
                                    'SeniorCitizen','PhoneService']).columns
target = ['Churn']
df_majority = encoded_df[encoded_df.Churn == 0]
df_minority = encoded_df[encoded_df.Churn == 1]


# Downsample majority class
df_majority_downsampled = resample(df_majority,
                                 replace=False,    # sample without replacement
                                 n_samples=down,     # to match minority class
                                 random_state=42) # reproducible results
# combine the new dataframes
df_downsampled = pd.concat([df_majority_downsampled, df_minority])

df_downsampled.Churn.value_counts()
y = df_downsampled[target]
X = df_downsampled[features]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50, random_state=42, stratify=y)
#LightGBM
clf = lgb.LGBMClassifier(drop_rate=0.9, min_data_in_leaf=800, max_bin=500,
                         n_estimators=5000, min_sum_hessian_in_leaf=1, importance_type='gain',
                         learning_rate=0.4, bagging_fraction=0.9, colsample_bytree=1.0,
                         feature_fraction=0.1, lambda_l1=5.0, lambda_l2=3.0, max_depth=9,
                         min_child_samples=55, min_child_weight=5.0, min_split_gain=0.1,
                         num_leaves=4000, subsample = 0.75)

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(f'Precision: {precision_score(y_test, y_pred)}')
print(f'Recall: {recall_score(y_test, y_pred)}')
print(f'Confusion Matrix: {confusion_matrix(y_test, y_pred)}')

fig, ax = plt.subplots(figsize = (10,10))
sns.set_style(style='whitegrid')
y_prob = clf.predict_proba(X_test)[::,1]
y_pred_proba = y_prob
fpr_rand, tpr_rand, _ = roc_curve(y_test,  y_pred_proba)
auc_rand = roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr_rand,tpr_rand,label="LightGBM, AUC="+str(auc_rand))
plt.legend(loc=4)
plt.title('ROC Curve')
plt.show()

feature_imp = pd.DataFrame(sorted(zip(clf.feature_importances_, X.columns)), columns=['Value','Feature'])
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False))
plt.title('LightGBM Features (avg over folds)')
plt.tight_layout()
plt.show()
encoded_df['extra_charges'] = (encoded_df.MonthlyCharges * encoded_df.tenure) - encoded_df.TotalCharges
encoded_df.head()
features = encoded_df.drop(columns=['Churn','tenure','customerID','gender','services','avg_price',
                                    'SeniorCitizen','PhoneService']).columns
target = ['Churn']
df_majority = encoded_df[encoded_df.Churn == 0]
df_minority = encoded_df[encoded_df.Churn == 1]


# Downsample majority class
df_majority_downsampled = resample(df_majority,
                                 replace=False,    # sample without replacement
                                 n_samples=down,     # to match minority class
                                 random_state=42) # reproducible results
# combine the new dataframes
df_downsampled = pd.concat([df_majority_downsampled, df_minority])

df_downsampled.Churn.value_counts()
y = df_downsampled[target]
X = df_downsampled[features]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50, random_state=42, stratify=y)
#LightGBM
clf = lgb.LGBMClassifier(drop_rate=0.9, min_data_in_leaf=800, max_bin=500,
                         n_estimators=5000, min_sum_hessian_in_leaf=1, importance_type='gain',
                         learning_rate=0.4, bagging_fraction=0.9, colsample_bytree=1.0,
                         feature_fraction=0.1, lambda_l1=5.0, lambda_l2=3.0, max_depth=9,
                         min_child_samples=55, min_child_weight=5.0, min_split_gain=0.1,
                         num_leaves=4000, subsample = 0.75)

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(f'Precision: {precision_score(y_test, y_pred)}')
print(f'Recall: {recall_score(y_test, y_pred)}')
print(f'Confusion Matrix: {confusion_matrix(y_test, y_pred)}')

fig, ax = plt.subplots(figsize = (10,10))
sns.set_style(style='whitegrid')
y_prob = clf.predict_proba(X_test)[::,1]
y_pred_proba = y_prob
fpr_rand, tpr_rand, _ = roc_curve(y_test,  y_pred_proba)
auc_rand = roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr_rand,tpr_rand,label="LightGBM, AUC="+str(auc_rand))
plt.legend(loc=4)
plt.title('ROC Curve')
plt.show()

feature_imp = pd.DataFrame(sorted(zip(clf.feature_importances_, X.columns)), columns=['Value','Feature'])
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False))
plt.title('LightGBM Features (avg over folds)')
plt.tight_layout()
plt.show()
features = encoded_df.drop(columns=['Churn','tenure','customerID','gender','services',
                                    'SeniorCitizen','PhoneService']).columns
target = ['Churn']
df_majority = encoded_df[encoded_df.Churn == 0]
df_minority = encoded_df[encoded_df.Churn == 1]


# Downsample majority class
df_majority_downsampled = resample(df_majority,
                                 replace=False,    # sample without replacement
                                 n_samples=down,     # to match minority class
                                 random_state=42) # reproducible results
# combine the new dataframes
df_downsampled = pd.concat([df_majority_downsampled, df_minority])

df_downsampled.Churn.value_counts()
y = df_downsampled[target]
X = df_downsampled[features]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50, random_state=42, stratify=y)
```
